# CRM Sales Opportunities Data Analysis
This notebook demonstrates data loading, cleaning, exploration, analysis, visualization, and insights for the CRM Sales Opportunities dataset.

## 1. Import Required Libraries & load data


```python
import pandas as pd
sales_pipeline = pd.read_csv("sales_pipeline.csv") 
accounts = pd.read_csv("accounts.csv") 
products = pd.read_csv("products.csv") 
sales_teams = pd.read_csv("sales_teams.csv")
```

```python
# Quick look at each table
for name, df in zip(['sales_pipeline', 'accounts', 'products', 'sales_teams'], [sales_pipeline, accounts, products, sales_teams]):
    print(f'--- {name} ---')
    print('Shape:', df.shape)
    display(df.head())
    print(df.info())
    display(df.describe(include='all'))
```

```python
# Check for duplicate rows in each table
for name, df in zip(['sales_pipeline', 'accounts', 'products', 'sales_teams'], [sales_pipeline, accounts, products, sales_teams]):
    print(f'--- {name} ---')
    print('Duplicate rows:', df.duplicated().sum())
    print()
```

- Typo 'technolgy' found in accounts.sector.

## 3. Clean the Data

### Convert Date Columns to Datetime
Convert `engage_date` and `close_date` columns in the sales pipeline to datetime format for accurate calculations.

```python
# Convert engage_date and close_date to datetime, and check for new NaT (missing/invalid dates)
date_cols = ['engage_date', 'close_date']
missing_before = sales_pipeline[date_cols].isna().sum()
sales_pipeline[date_cols] = sales_pipeline[date_cols].apply(pd.to_datetime, errors='coerce')
missing_after = sales_pipeline[date_cols].isna().sum()
missing_report = pd.DataFrame({
    'missing_before': missing_before,
    'missing_after': missing_after,
    'new_missing_after_conversion': missing_after - missing_before
})
display(missing_report)
```

### Normalize Numeric Columns
Convert `close_value` and `year_established` to integer-like values for aggregation consistency.

```python
# Normalize numeric types for aggregation
sales_pipeline['close_value'] = pd.to_numeric(sales_pipeline['close_value'], errors='coerce').round(0).astype('Int64')
accounts['year_established'] = pd.to_numeric(accounts['year_established'], errors='coerce').round(0).astype('Int64')
```

### Fix Typo in Sector Column
Correct the typo 'technolgy' to 'technology' in the `sector` column of the accounts table.

```python
# Fix typo in sector column
accounts['sector'] = accounts['sector'].replace('technolgy', 'technology')
```

## 4. Validate and Document Missing Values
Check and document missing values in important columns before merging. These are the join keys and core deal attributes used later in analysis: `account`, `opportunity_id`, `product`, `deal_stage`, `sales_agent`, and `subsidiary_of`.

```python
# Check missing values in core join keys and deal attributes
key_columns = {
    'opportunity_id': sales_pipeline['opportunity_id'],
    'account': sales_pipeline['account'],
    'product': sales_pipeline['product'],
    'deal_stage': sales_pipeline['deal_stage'],
    'sales_agent': sales_pipeline['sales_agent'],
    'subsidiary_of': accounts['subsidiary_of']
}
for name, col in key_columns.items():
    missing_pct = col.isna().mean() * 100
    print(f"Missing '{name}': {missing_pct:.1f}%")

print("\nNotes:")
print("- Missing 'opportunity_id' or 'product' would break key deal identification and joins.")
print("- Missing 'deal_stage' or 'sales_agent' would weaken analysis and should be investigated.")
print("- Missing 'account' may be acceptable for unassigned or early-stage deals, but it is still documented and reviewed below for active deals with engage_date.")
print("- Missing 'subsidiary_of' is expected for independent companies.")
```

### Review active opportunities missing account
These opportunities have a start date, so they appear active, but they are not linked to an account. They should be reviewed to validate and complete the data before analysis.

```python
# Review active opportunities that are missing an account
account_blank = sales_pipeline[sales_pipeline['account'].isna() & sales_pipeline['engage_date'].notna()]
count = len(account_blank)
print(f"Blank-account rows with engage_date: {count}")
if count > 0:
    print("Sample of active opportunities missing account:")
    display(account_blank[['opportunity_id', 'product', 'deal_stage', 'sales_agent', 'engage_date', 'close_value']].head(10))
    print("These rows should be reviewed to validate the data and complete missing account information.")
else:
    print("No active opportunities with engage_date are missing account.")
```

## 5. Merge All Tables into a Master DataFrame
Join the four tables on their respective keys to create a single master dataframe for analysis.

```python
# Merge sales_pipeline with accounts
master_df = sales_pipeline.merge(accounts, on='account', how='left')
# Merge with products
master_df = master_df.merge(products, on='product', how='left')
# Merge with sales_teams
master_df = master_df.merge(sales_teams, on='sales_agent', how='left')
print('Master dataframe shape:', master_df.shape)
```

## 6. Add Calculated Columns
Add useful columns:
- `deal_duration_days`: Days between engage and close date (if both exist)
- `quarter`: Quarter of the engage date
- `win_flag`: 1 if deal_stage is 'Won', else 0

```python
# Calculate deal_duration_days
master_df['deal_duration_days'] = (master_df['close_date'] - master_df['engage_date']).dt.days
# Extract quarter from engage_date
master_df['quarter'] = master_df['engage_date'].dt.to_period('Q')
# Create win_flag column
master_df['win_flag'] = (master_df['deal_stage'] == 'Won').astype(int)
```

```python
# Calculate total and average value of won deals
total_revenue = master_df.loc[master_df['win_flag'] == 1, 'close_value'].sum()
avg_deal_value = master_df.loc[master_df['win_flag'] == 1, 'close_value'].mean()
print(f"Total revenue from won deals: {total_revenue:,.0f}")
print(f"Average deal value (won deals): {avg_deal_value:,.0f}")
```

## 8. Exploratory Data Analysis
Explore patterns, distributions, and data quality after cleaning.

```python
# EDA: Distributions and key metrics
import matplotlib.pyplot as plt
import seaborn as sns

# Compare won vs lost deals (most business-relevant)
won = master_df[master_df['win_flag'] == 1]
lost = master_df[master_df['deal_stage'] == 'Lost']

# Show won deal distribution 
plt.figure(figsize=(10, 4))
plt.hist(won['close_value'].dropna(), bins=40, edgecolor='black', color='green', alpha=0.7)
plt.yscale('log')
plt.title(f'Won close_value Distribution ({len(won)} deals, log scale)')
plt.xlabel('close_value')
plt.ylabel('frequency (log)')
plt.tight_layout()
plt.show()

# Summary stats
print('=== WON DEALS ===')
print(won['close_value'].describe().round(2))
print('\n=== LOST DEALS ===')
print(f"count: {len(lost)}")
print(f"mean: {lost['close_value'].mean():.2f}")
print('Note: lost close_value is effectively zero for this segment, so the full describe output is not needed.')

print('\n=== DEAL STAGE COUNTS ===')
stage_counts = master_df['deal_stage'].value_counts()
stage_pct = (stage_counts / stage_counts.sum() * 100).round(2)
for stage, count in stage_counts.items():
    print(f"{stage}: {count} deals ({stage_pct[stage]}%)")

print('\n=== IMPORTANT MISSING VALUES IN master_df (%) ===')
important_cols = ['opportunity_id', 'product', 'deal_stage', 'sales_agent', 'account']
missing = master_df[important_cols].isna().mean() * 100
print(missing.round(2))
```

## 9. Business Question 1: 
### How is each sales team performing compared to the rest?
Analyze total and won deal values by sales team, and visualize the results.

```python
# Group by regional_office for team performance
team_perf = master_df.groupby('regional_office').agg(total_deals=('opportunity_id','count'),
                                                        won_deals=('win_flag','sum'),
                                                        total_value=('close_value','sum'),
                                                        close_value=('close_value',lambda x: x[master_df.loc[x.index,'win_flag']==1].sum()))
team_perf['win_rate'] = team_perf['won_deals'] / team_perf['total_deals']
team_perf_sorted = team_perf.sort_values('close_value')
display(team_perf_sorted)
plt.figure(figsize=(12,5))
# .reset_index() is needed to turn regional_office back into a column for seaborn
team_perf_plot = team_perf_sorted.reset_index()
sns.barplot(data=team_perf_plot, x='regional_office', y='close_value', hue='regional_office', palette='viridis', legend=False)
for idx, row in team_perf_plot.iterrows():
    plt.text(idx, row['close_value'] * 1.01, f"{row['close_value']:,}", ha='center', va='bottom', fontsize=10)
plt.title('Total Won Close_Value by Sales Team (Regional Office)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight:** The table and chart above show which sales teams (based on the **regional offices**) are leading in total won deal value and win rate. West regional office has the most number. 

## 10. Business Question 2: 
### Are any sales agents lagging behind?
Analyze performance by sales agent and visualize the lowest performers.

```python
# Agent performance: show bottom 10 by close_value
agent_perf = master_df.groupby('sales_agent').agg(total_deals=('opportunity_id','count'),
                                                      won_deals=('win_flag','sum'),
                                                      total_value=('close_value','sum'),
                                                      close_value=('close_value',lambda x: x[master_df.loc[x.index,'win_flag']==1].sum()))
agent_perf['win_rate'] = agent_perf['won_deals'] / agent_perf['total_deals']
display(agent_perf.sort_values('close_value').head(10))
plt.figure(figsize=(12,5))
sns.barplot(data=agent_perf.sort_values('close_value').head(10).reset_index(), x='sales_agent', y='close_value', hue='sales_agent', palette='Blues')
for idx, row in agent_perf.sort_values('close_value').head(10).reset_index().iterrows():
    plt.text(idx, row['close_value'] * 1.01, f"{row['close_value']:,}", ha='center', va='bottom', fontsize=10)
plt.title('Bottom 10 Sales Agents by Won Close_Value') 
plt.xticks(rotation=45)
plt.show()
```

**Insight:** The table and chart above highlight the sales agents with the lowest won deal values, helping identify those who may need support or training.

## 11. Business Question 3: 
### Are there any quarter-over-quarter trends?
Analyze and visualize total won and lost value plus win/lost rates by quarter.

```python
# Quarter-over-quarter analysis
import matplotlib.pyplot as plt
quarterly = master_df.groupby('quarter').agg(
    total_deals=('opportunity_id','count'),
    won_deals=('win_flag','sum'),
    lost_deals=('deal_stage', lambda x: (x == 'Lost').sum())
)
quarterly['close_value'] = master_df.loc[master_df['win_flag'] == 1].groupby('quarter')['close_value'].sum()
quarterly['close_value'] = quarterly['close_value'].fillna(0)
quarterly['win_rate'] = quarterly['won_deals'] / quarterly['total_deals']
quarterly['lost_rate'] = quarterly['lost_deals'] / quarterly['total_deals']

quarterly_won = quarterly[['close_value', 'won_deals', 'win_rate']].copy()
quarterly_lost = quarterly[['lost_deals', 'lost_rate']].copy()

print('Won deals by quarter:')
display(quarterly_won)
print('Lost deals by quarter:')
display(quarterly_lost)

quarterly_plot = quarterly.reset_index()
quarterly_plot['quarter'] = quarterly_plot['quarter'].astype(str)

fig, ax1 = plt.subplots(figsize=(10,5))
bars = ax1.bar(quarterly_plot['quarter'], quarterly_plot['close_value'], color='mediumseagreen', width=0.6)
ax1.set_ylabel('close_value')
ax1.set_xlabel('quarter')
ax1.set_title('Quarter-over-Quarter Won Close Value and Win Rate')
ax1.tick_params(axis='x', rotation=0)

for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height + max(quarterly_plot['close_value']) * 0.02,
        f"{int(height):,}",
        ha='center',
        va='bottom',
        fontsize=9,
    )

ax2 = ax1.twinx()
ax2.plot(quarterly_plot['quarter'], quarterly_plot['win_rate'], color='darkblue', marker='o', linewidth=2)
ax2.set_ylabel('Win Rate')
for quarter, win_rate in zip(quarterly_plot['quarter'], quarterly_plot['win_rate']):
    ax2.text(quarter, win_rate + 0.02, f"{win_rate:.0%}", ha='center', va='bottom', color='darkblue', fontsize=9)

ax1.legend(['close_value'], loc='upper left')
ax2.legend(['win rate'], loc='upper right')

plt.tight_layout()
plt.show()
```

**Insight:** Revenue from won deals has grown even though the win rate has declined. This can happen because win rate measures deal success by count (`won_deals / total_deals`), while `close_value` measures the total value of closed-won opportunities. In short, the business may be winning fewer opportunities, but the opportunities that are won are higher value.

## 12. Business Question 4: 
### Do any products have better win rates?
Analyze win rates and won close_value by product, and visualize the results.

```python
# Product win rates and value
product_perf = master_df.groupby('product').agg(total_deals=('opportunity_id','count'),
                                                    won_deals=('win_flag','sum'),
                                                    close_value=('close_value',lambda x: x[master_df.loc[x.index,'win_flag']==1].sum()))
product_perf['win_rate'] = product_perf['won_deals'] / product_perf['total_deals']
product_perf_sorted = product_perf.sort_values('win_rate', ascending=True)
display(product_perf_sorted)
fig, ax1 = plt.subplots(figsize=(10,5))
sns.barplot(data=product_perf_sorted.reset_index(), x='product', y='win_rate', ax=ax1, color='lightblue')
for bar in ax1.patches:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height / 2,
        f"{height:.0%}",
        ha='center',
        va='center',
        color='navy',
        fontsize=10,
        fontweight='normal'
    )
ax1.set_ylabel('Win Rate')
ax1.set_title('Win Rate by Product')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight:** The table and chart above show which products have the highest win rates, helping to identify top performers and potential focus areas.
