# CRM Sales Opportunities Data Analysis
This notebook demonstrates data loading, cleaning, exploration, analysis, visualization, and insights for the CRM Sales Opportunities dataset.

## 1. Import Required Libraries & load data


```python
import pandas as pd
sales_pipeline = pd.read_csv("sales_pipeline.csv") 
accounts = pd.read_csv("accounts.csv") 
products = pd.read_csv("products.csv") 
sales_teams = pd.read_csv("sales_teams.csv")
```

## 3. Clean the Data

### Convert Date Columns to Datetime
Convert `engage_date` and `close_date` columns in the sales pipeline to datetime format for accurate calculations.

```python
# Convert engage_date and close_date to datetime, and check for new NaT (missing/invalid dates)
date_cols = ['engage_date', 'close_date']
missing_before = sales_pipeline[date_cols].isna().sum()
sales_pipeline[date_cols] = sales_pipeline[date_cols].apply(pd.to_datetime, errors='coerce')
missing_after = sales_pipeline[date_cols].isna().sum()
missing_report = pd.DataFrame({
    'missing_before': missing_before,
    'missing_after': missing_after,
    'new_missing_after_conversion': missing_after - missing_before
})
display(missing_report)
```

### Normalize Numeric Columns
Convert `close_value` and `year_established` to integer-like values for aggregation consistency.

```python
# Normalize numeric types for aggregation
sales_pipeline['close_value'] = pd.to_numeric(sales_pipeline['close_value'], errors='coerce')
accounts['year_established'] = pd.to_numeric(accounts['year_established'], errors='coerce').round(0).astype('Int64')
```

### Fix Typo in Sector Column
Correct the typo 'technolgy' to 'technology' in the `sector` column of the accounts table.

```python
# Fix typo in sector column
accounts['sector'] = accounts['sector'].replace('technolgy', 'technology')
```

## 4. Validate and Document Missing Values
Check and document missing values in important columns before merging. These are the join keys and core deal attributes used later in analysis: `account`, `opportunity_id`, `product`, `deal_stage`, `sales_agent`, and `subsidiary_of`.

```python
# Check missing values in core join keys and deal attributes
key_columns = {
    'opportunity_id': sales_pipeline['opportunity_id'],
    'account': sales_pipeline['account'],
    'product': sales_pipeline['product'],
    'deal_stage': sales_pipeline['deal_stage'],
    'sales_agent': sales_pipeline['sales_agent'],
    'subsidiary_of': accounts['subsidiary_of']
}
for name, col in key_columns.items():
    missing_pct = col.isna().mean() * 100
    print(f"Missing '{name}': {missing_pct:.1f}%")

print("\nNotes:")
print("- Missing 'opportunity_id' or 'product' would break key deal identification and joins.")
print("- Missing 'deal_stage' or 'sales_agent' would weaken analysis and should be investigated.")
print("- Missing 'account' may be acceptable for unassigned or early-stage deals, but it is still documented and reviewed below for active deals with engage_date.")
print("- Missing 'subsidiary_of' is expected for independent companies.")
```

### Review active opportunities missing account
These opportunities have a start date, so they appear active, but they are not linked to an account. They should be reviewed to validate and complete the data before analysis.

```python
# Review active opportunities that are missing an account
account_blank = sales_pipeline[sales_pipeline['account'].isna() & sales_pipeline['engage_date'].notna()]
count = len(account_blank)
print(f"Blank-account rows with engage_date: {count}")
if count > 0:
    print("Sample of active opportunities missing account:")
    display(account_blank[['opportunity_id', 'product', 'deal_stage', 'sales_agent', 'engage_date', 'close_value']].head(10))
    print("These rows should be reviewed to validate the data and complete missing account information.")
else:
    print("No active opportunities with engage_date are missing account.")
```

## 5. Merge All Tables into a Master DataFrame
Join the four tables on their respective keys to create a single master dataframe for analysis.

```python
# Merge sales_pipeline with accounts
master_df = sales_pipeline.merge(accounts, on='account', how='left')
# Merge with products
master_df = master_df.merge(products, on='product', how='left')
# Merge with sales_teams
master_df = master_df.merge(sales_teams, on='sales_agent', how='left')
print('Master dataframe shape:', master_df.shape)
```

## 6. Add Calculated Columns
Add useful columns:
- `deal_duration_days`: Days between engage and close date (if both exist)
- `quarter`: Quarter of the engage date
- `win_flag`: 1 if deal_stage is 'Won', else 0

```python
# Calculate deal_duration_days
master_df['deal_duration_days'] = (master_df['close_date'] - master_df['engage_date']).dt.days
# Extract quarter from engage_date
master_df['quarter'] = master_df['engage_date'].dt.to_period('Q')
# Create win_flag column
master_df['win_flag'] = (master_df['deal_stage'] == 'Won').astype(int)
```

## 8. Exploratory Data Analysis
Explore patterns, distributions, and data quality after cleaning.

```python
# EDA: Distributions and key metrics
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

won = master_df[master_df['win_flag'] == 1]

fig, ax = plt.subplots(figsize=(12, 5))

# Plot histogram and capture bar data
counts, edges, patches = ax.hist(
    won['close_value'].dropna(), 
    bins=40, 
    edgecolor='black', 
    color='green', 
    alpha=0.7
)

# Add data labels on bars (only show if count > 0)
for count, edge, patch in zip(counts, edges, patches):
    if count > 0:
        ax.text(
                patch.get_x() + patch.get_width() / 2,
                patch.get_height() + 10,
                f'{int(count)}',
                ha='center', va='bottom',
                fontsize=7, rotation=90,
                fontweight='bold'
)

# X axis in K format
ax.xaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f'${int(x/1000)}K' if x >= 1000 else f'${int(x)}')
)

# Y axis plain numbers
ax.yaxis.set_major_formatter(
    mticker.FuncFormatter(lambda y, _: f'{int(y):,}')
)

ax.set_title(f'Won Revenue Distribution ({len(won)} deals)')
ax.set_xlabel('Deal Value')
ax.set_ylabel('Number of Deals')
plt.tight_layout()
plt.show()
```

## 9. Business Question 1: 
### How is each sales team performing compared to the rest?
Analyze total and won deal values by sales team, and visualize the results.

```python
# Group by regional_office for team performance
import seaborn as sns

team_perf = master_df.groupby('regional_office').agg(
    total_deals=('opportunity_id', 'count'),
    won_deals=('win_flag', 'sum'),
    total_revenue=('close_value', lambda x: x[master_df.loc[x.index, 'win_flag'] == 1].sum())
)
team_perf['win_rate'] = team_perf['won_deals'] / team_perf['total_deals']
team_perf_sorted = team_perf.sort_values('total_revenue')

# Display table
team_perf_display = team_perf_sorted.copy()
team_perf_display['total_deals'] = team_perf_display['total_deals'].apply(lambda x: f'{x:,}')
team_perf_display['won_deals'] = team_perf_display['won_deals'].apply(lambda x: f'{x:,}')
team_perf_display['total_revenue'] = team_perf_display['total_revenue'].apply(lambda x: f'${x/1_000_000:.2f}M')
team_perf_display['win_rate'] = team_perf_display['win_rate'].apply(lambda x: f'{x:.1%}')
display(team_perf_display)

# Bar chart
team_perf_plot = team_perf_sorted.reset_index()
plt.figure(figsize=(12, 5))
sns.barplot(data=team_perf_plot, x='regional_office', y='total_revenue',
            hue='regional_office', palette='viridis', legend=False)

# Data labels
for idx, row in team_perf_plot.iterrows():
    plt.text(idx, row['total_revenue'] * 1.01,
             f"${row['total_revenue']/1_000_000:.2f}M",
             ha='center', va='bottom', fontsize=10)

# Y axis in M format
plt.gca().yaxis.set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'${y/1_000_000:.1f}M')
)

plt.title('Total Won Revenue by Sales Team (Regional Office)')
plt.xlabel('Regional Office')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insight:** This section shows which sales teams (based on the **regional offices**) are leading in total revenue and win rate.  
The West region leads in total revenue at $3.57M from 1,438 won deals, but the East region has the highest win rate at 51.1% — meaning East converts more deals successfully even with fewer total deals. Central has the most deal volume (3,512) but the lowest win rate (46.4%), suggesting possible issues with lead qualification or deal selection.

## 10. Business Question 2: 
### Are any sales agents lagging behind?
Analyze performance by sales agent and visualize the 10 lowest performers.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agent performance: bottom 10 by revenue (won deals only)
closed = master_df[master_df['deal_stage'].isin(['Won', 'Lost'])]
agent_perf = closed.groupby('sales_agent').agg(
    total_deals=('opportunity_id', 'count'),
    won_deals=('win_flag', 'sum'),
    revenue=('close_value', lambda x: x[master_df.loc[x.index, 'win_flag'] == 1].sum())
)
agent_perf['win_rate'] = agent_perf['won_deals'] / agent_perf['total_deals']
agent_perf_sorted = agent_perf.sort_values('revenue')
bottom10 = agent_perf_sorted.head(10)

# Display table with clean formatting
bottom10_display = bottom10.copy()
bottom10_display['total_deals'] = bottom10_display['total_deals'].apply(lambda x: f'{x:,}')
bottom10_display['won_deals'] = bottom10_display['won_deals'].apply(lambda x: f'{x:,}')
bottom10_display['revenue'] = bottom10_display['revenue'].apply(lambda x: f'${x/1_000:.0f}K')
bottom10_display['win_rate'] = bottom10_display['win_rate'].apply(lambda x: f'{x:.1%}')
display(bottom10_display)

# Bar chart
bottom10_plot = bottom10.reset_index()
plt.figure(figsize=(12, 5))
sns.barplot(data=bottom10_plot, x='sales_agent', y='revenue',
            hue='sales_agent', palette='viridis', legend=False)

# Data labels
for idx, row in bottom10_plot.iterrows():
    plt.text(idx, row['revenue'] * 1.01,
             f"${row['revenue']/1_000:.0f}K",
             ha='center', va='bottom', fontsize=9)

# Y axis in K format
plt.gca().yaxis.set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'${y/1_000:.0f}K')
)

plt.title('Bottom 10 Agents by Won Revenue')
plt.xlabel('Sales Agent')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insight:** The table and chart above highlight the sales agents with the 10 lowest total revenue. The bar chart makes it clear that Violet Mclelland has the smallest won value at $123K and Wilburn Farren is also far below peers with $157K from only 55 won deals. Most bottom performers still maintain win rates above 60%, suggesting the issue is pipeline size and deal volume rather than closing ability.


## 11. Business Question 3: 
### Are there any quarter-over-quarter trends?
Analyze and visualize total won and lost value plus win/lost rates by quarter.

```python
# Chart
quarterly_plot = quarterly.reset_index()
quarterly_plot['quarter'] = quarterly_plot['quarter'].astype(str)

fig, ax1 = plt.subplots(figsize=(10, 5))
bars = ax1.bar(quarterly_plot['quarter'], quarterly_plot['revenue'],
               color='mediumseagreen', width=0.6)

# Y axis formatter
ax1.yaxis.set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'${y/1_000:.0f}K' if y < 1_000_000 else f'${y/1_000_000:.1f}M')
)
ax1.set_ylabel('Won Revenue')
ax1.set_xlabel('Quarter')
ax1.set_title('Quarter-over-Quarter Won Revenue and Win Rate')
ax1.tick_params(axis='x', rotation=0)

# Bar data labels
for bar in bars:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height + max(quarterly_plot['revenue']) * 0.02,
        f'${height/1_000:.0f}K' if height < 1_000_000 else f'${height/1_000_000:.2f}M',
        ha='center', va='bottom', fontsize=9
    )

# Win rate line
ax2 = ax1.twinx()
ax2.plot(quarterly_plot['quarter'], quarterly_plot['win_rate'],
         color='darkblue', marker='o', linewidth=2)
ax2.set_ylabel('')
ax2.yaxis.set_visible(False)  
for quarter, win_rate in zip(quarterly_plot['quarter'], quarterly_plot['win_rate']):
    ax2.text(quarter, win_rate + 0.02, f'{win_rate:.1%}',
             ha='center', va='bottom', color='darkblue', fontsize=9)

ax1.legend(['Won Revenue'], loc='lower left')
ax2.legend(['Win Rate'], loc='lower right')

plt.tight_layout()
plt.show()
```

**Insight:** Won revenue increased strongly from $524K in 2016Q4 to its highest point of $2.97M in 2017Q2. This happened even though the win rate decreased from 69.8% to 52.2% during the same period. This means the business was winning fewer deals, but the deals it won were larger and brought in more revenue. After that, revenue decreased slightly to $2.79M in 2017Q3, while the win rate also dropped to its lowest level at 43.6%. This period may show a weaker sales quarter, because both revenue and win rate went down together. In 2017Q4, revenue recovered to $1.55M. Overall, the pattern suggests that the company moved toward higher-value deals over time, but 2017Q3 should be investigated further to understand why performance dropped.

## 12. Business Question 4: 
### Do any products have better win rates?
Analyze win rates and won revenue by product, and visualize the results.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Product win rates and revenue
product_perf = master_df.groupby('product').agg(
    total_deals=('opportunity_id', 'count'),
    won_deals=('win_flag', 'sum'),
    revenue=('close_value', lambda x: x[master_df.loc[x.index, 'win_flag'] == 1].sum())
)
product_perf['win_rate'] = product_perf['won_deals'] / product_perf['total_deals']
product_perf_sorted = product_perf.sort_values('win_rate', ascending=True)

# Display table with clean formatting
product_perf_display = product_perf_sorted.copy()
product_perf_display['total_deals'] = product_perf_display['total_deals'].apply(lambda x: f'{x:,}')
product_perf_display['won_deals'] = product_perf_display['won_deals'].apply(lambda x: f'{x:,}')
product_perf_display['revenue'] = product_perf_display['revenue'].apply(
    lambda x: f'${x/1_000:.0f}K' if x < 1_000_000 else f'${x/1_000_000:.2f}M')
product_perf_display['win_rate'] = product_perf_display['win_rate'].apply(lambda x: f'{x:.1%}')
display(product_perf_display)

# Bar chart
fig, ax1 = plt.subplots(figsize=(10, 5))
sns.barplot(data=product_perf_sorted.reset_index(), x='product', y='win_rate',
            ax=ax1, color='lightblue')

# Data labels
for bar in ax1.patches:
    height = bar.get_height()
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        height / 2,
        f'{height:.1%}',
        ha='center', va='center',
        color='navy', fontsize=10
    )

# Y axis as percentage
ax1.yaxis.set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'{y:.0%}')
)

ax1.set_ylabel('Win Rate')
ax1.set_xlabel('Product')
ax1.set_title('Win Rate by Product')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Insight:** The table and chart above show which products have the highest win rates, helping to identify top performers and potential focus areas.

GTX Plus Pro has the highest win rate at 49.5%, followed by GTXPro at 49.3% and GTX Basic at 49.0%. These three products perform almost the same. GTK 500 has the lowest win rate at 37.5%, which is clearly below the other products. Most of the other products are between 46% and 50%, so product choice does not seem to strongly affect win success, except for GTK 500. GTK 500 may need pricing, positioning, or target audience review to understand why it underperforms compared to all other products.

## Executive Summary

This analysis examined 8,800 B2B sales pipeline records across four 
dimensions: team performance, agent performance, quarterly trends, 
and product win rates. Below are the key findings.

---

**1. The West region generates the most revenue — but East is the 
most efficient.**
West leads with $3.57M in won revenue, while East has the highest 
win rate at 51.1% despite fewer total deals. Central processes the 
most deals (3,512) but has the lowest win rate (46.4%), suggesting 
possible issues.

**2. Low-revenue agents have a pipeline problem, not a closing problem.**
The bottom 10 agents by revenue mostly maintain win rates above 60% 
on closed deals. Wilburn Farren has the fewest won deals (55) despite 
a 69.6% win rate — indicating low activity rather than poor skill. 
These agents need more pipeline, not more training.

**3. Revenue is growing even as win rate declines — a positive signal.**
Won revenue peaked at $2.97M in 2017Q2, up from $524K in 2016Q4, 
despite the win rate falling from 69.8% to 52.2%. This suggests the 
business is closing fewer but higher-value deals over time. The 
exception is 2017Q3, where both revenue and win rate dropped together, 
indicating a difficult sales period worth investigating.

**4. Product choice has minimal impact on win rate — except GTK 500.**
Six of seven products cluster tightly between 46% and 50% win rate. 
GTK 500 is the clear outlier at 37.5%, more than 10 points below the 
next product. GTXPro generates the highest revenue at $3.51M despite 
not having the top win rate, making it the most valuable product overall.
